1.Write a for loop to print the following pattern using nested loops:

1
12
123
1234
12345


```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end='')
    print()
```

2.Given the dictionary below, convert each student’s marks into grades based on the following criteria:

80–100 → A

70–79 → B

Below 70 → C
marks = {"Alice": 85, "Bob": 72, "Charlie": 90, "David": 67}


```python
marks = {"Alice": 85, "Bob": 72, "Charlie": 90, "David": 67}
grades = {}

for name, score in marks.items():
    if score >= 80:
        grades[name] = 'A'
    elif score >= 70:
        grades[name] = 'B'
    else:
        grades[name] = 'C'

print(grades)
```

3.Write a for loop to find all numbers between 1 and 30 that are divisible by 3 or 5, but not both.

```python
result = []

for i in range(1, 31):
    if (i % 3 == 0 or i % 5 == 0) and not (i % 3 == 0 and i % 5 == 0):
        result.append(i)

print(result)
```

4.Write a function that takes a list of fruits and prints each fruit with its index number.

```python
def print_fruits(fruits):
    for index, fruit in enumerate(fruits, start=1):
        print(f"Fruit {index}: {fruit}")

# Test
fruits = ["apple", "banana", "cherry", "date"]
print_fruits(fruits)
```

5.Write a function that takes two lists — one of names and one of scores — and prints them together.

```python
def print_scores(names, scores):
    for name, score in zip(names, scores):
        print(f"{name} scored {score}")

# Test
names = ["Aman", "Priya", "Rahul"]
scores = [89, 95, 72]
print_scores(names, scores)
```
