```python
#1 . inheritance
class A:
  def feat1(self):
    print('feature 1 working')
  def feat2(self):
    print('feature 2 working')
  def feat3(self):
    print('feature 3 working')
a1 = A()
a2 = A()
a1.feat1()
a1.feat2()
a1.feat3()
```

```python
 # single level inheritance.

class A: #super class\parent class
  def feat1(self):
    print('feature 1 working')
  def feat2(self):
    print('feature 2 working')
  def feat3(self):
    print('feature 3 working')

class B(A): # sub class\child class
  def feat4(self):
    print('feature 4 working')
  def feat5(self):
    print('feature 5 working')
a1 = A()
b1 = B()
a1.feat1()
b1.feat1()

```

```python
# multi-level inheritance
class A: #super class\parent class
  def feat1(self):
    print('feature 1 working')
  def feat2(self):
    print('feature 2 working')
  def feat3(self):
    print('feature 3 working')


class B(A): # sub class\child class
  def feat4(self):
    print('feature 4 working')
  def feat5(self):
    print('feature 5 working')

class C(B):  # sub class\child clas
  def feat6(self):
    print('feature 6 working')
  def feat7(self):
    print('feature 7 working')

a1 = A()
b1 = B()
c1 = C()
a1.feat1()
a1.feat2()
b1.feat1()
c1.feat6()
```

```python
# multiple inheritance
class A: #super class\parent class
  def feat1(self):
    print('feature 1 working')
  def feat2(self):
    print('feature 2 working')
  def feat3(self):
    print('feature 3 working')


class B(A): # sub class\child class
  def feat4(self):
    print('feature 4 working')
  def feat5(self):
    print('feature 5 working')

class C(B,A):  # sub class\child clas
  def feat6(self):
    print('feature 6 working')
  def feat7(self):
    print('feature 7 working')

a1 = A()
b1 = B()
c1 = C()
b1.feat1()
c1.feat1()
c1.feat6()
```

```python
class A:
  def __init__(self):
    print("in init block of A")
  def feat1(self):
    print("feature 1 working")
class B(A):
   def __init__(self):
    print("in init block of B")
   def feat2(self):
    super().__init__()
    print("feature 2 working")
a = B()
```

```python
class A:
  def __init__(self):
    print("in init block of A")

  def feat1(self):
    print("feature 1 working")

class B:
  def __init__(self):
    print("in init block of B")
    super().__init__()
  def feat2(self):
    print("feature 2 working")
class C(A,B): # Changed the order of base classes
  def feat3(self):
    print("feature 3 working")

a1 = C()
```

**SECOND PILLER OF OPPS**
**POLYMORPHISM**
(**for different class if there be same method performing task it called ploymorphism**)

```python
class laptop:
  def wap(self,ide):
    ide.excute()
class en1:
  def excute(self):
    print('compiling')
class en2:
  def excute(self):
    print('running')
ide = laptop()
ide.wap(en1())
ide.wap(en2())
```

```python
class laptop:
  def wap(self,ide):
    ide.excute()
class en1:
  def excute(self):
    print('compiling')
class en2:
  def excute(self):
    print('running')
ide=en2()
lap1=laptop()
lap1.wap(ide)
```

```python
class env:
  def excute(self):
    print('compiling')
class env1:
  def excute(self):
    print('running')
    print('grammer check')
class env2:
  def excute(self):
    print('compiling')
    print('running')
    print('grammer check')
class laptop:
  def code(self,ide):
    ide.excute()
ide = env2()
lap1 = laptop()
lap1.code(ide)
```
