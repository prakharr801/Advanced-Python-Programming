```python
class env:
  def excute(self):
    print('compiling')
class env1:
  def excute(self):
    print('running')
    print('grammer check')
class env2:
  def excute(self):
    print('compiling')
    print('running')
    print('grammer check')
class laptop:
  def code(self,ide):
    ide.excute()
ide = env2()
lap1 = laptop()
lap1.code(ide)
```

Encapsulation

```python
# public/global acces
class Students:
  def __init__(self,name,marks):
    self.marks = marks #public variable
    self.name = name   #public variable
  def display(self):   #public method
    print(self.name,self.marks)
s = Students('Abishek',90)
print(s.name)     #global acces allowed
print(s.marks)
s.display()
```
